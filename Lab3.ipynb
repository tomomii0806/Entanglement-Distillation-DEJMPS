## 1. Convergence Behavior Graphs  
Purpose: Show how each optimization method improves fidelity over iterations.  
Graph Type: Line Plot (Fidelity vs. Iterations)

```python
import matplotlib.pyplot as plt

iterations = list(range(1, 11))  # Example iterations
fidelity_gd = [0.70, 0.74, 0.77, 0.788] + [0.788] * 6
fidelity_ga = [0.70, 0.72, 0.75, 0.78, 0.787, 0.788, 0.788, 0.788, 0.788, 0.788]
fidelity_sa = [0.70, 0.73, 0.76, 0.78, 0.788, 0.788, 0.788, 0.788, 0.788, 0.788]

plt.plot(iterations, fidelity_gd, label="Gradient Descent", marker="o")
plt.plot(iterations, fidelity_ga, label="Genetic Algorithm", marker="s")
plt.plot(iterations, fidelity_sa, label="Simulated Annealing", marker="d")

plt.xlabel("Iterations")
plt.ylabel("Fidelity")
plt.title("Convergence of Optimization Algorithms")
plt.legend()
plt.grid()
plt.show()
```

## 2. Computational Efficiency Comparison  
Purpose: Show time complexity and performance across different algorithms.  
Graph Type: Bar Chart (Execution Time vs. Algorithm)

```python
algorithms = ["Gradient Descent", "Genetic Algorithm", "Simulated Annealing", "Grover's Search", "Multi-Round DEJMPS"]
times = [33.8, 528, 154, 12.1, 1.46]  # Example times in seconds

plt.bar(algorithms, times, color=["blue", "green", "red", "purple", "orange"])
plt.xlabel("Optimization Algorithm")
plt.ylabel("Time Taken (seconds)")
plt.title("Computational Efficiency Comparison")
plt.xticks(rotation=15)
plt.show()
```
